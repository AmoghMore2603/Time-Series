```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-car-sales.csv')
df
```

```python
df.rename(columns = {'Month':'ds','Sales':'y'},inplace = True)
```

```python
df
```

```python
df.info()
```

```python
df['ds'] = pd.to_datetime(df['ds'])
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8)) 
plt.scatter(df['ds'],df['y'])
```

```python
import fbprophet
model = fbprophet.Prophet()
model.fit(df)
```

```python
forecast = model.make_future_dataframe(periods = 17, freq= 'M',include_history = True)
forecast = model.predict(forecast)
```

```python
model.plot(forecast)
plt.show()
```

```python
import joblib
```

```python
joblib.dump(model, 'model_project')
```
